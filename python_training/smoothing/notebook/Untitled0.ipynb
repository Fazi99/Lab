```python
def image_make(wEVI, EVI):
    wdoy = int(wEVI.split('.')[2]) + 1
    tstr = "2005 %s" % (wdoy)
    wdate = datetime.datetime.strptime(tstr, '%Y %j')
    warray = np.load(wEVI).astype(np.float32) / 10000
    fig = plt.figure(figsize=(15,5), dpi=150)
    fig.add_subplot(1, 2, 1)
    plt.imshow(warray.reshape(500, 500), interpolation='nearest', vmin=0, vmax=1., cmap=plt.cm.Greens)
    plt.title(wdate.strftime('%Y.%m.%d'))
    plt.colorbar()
    fig.add_subplot(1, 2, 2)
    tstr = EVI.split('.')[1].lstrip('A')
    edoy = datetime.datetime.strptime(tstr, '%Y%j')
    image = gdal.Open(EVI, gdal.GA_ReadOnly)
    earray = image.ReadAsArray()
    plt.imshow(earray, interpolation='nearest', vmin=0, vmax=1., cmap=plt.cm.Greens)
    plt.title(edoy.strftime('%Y.%m.%d'))
    plt.colorbar()
    #plt.show()
    imgname = "../../img/MOD.EVI.%03d.wavelet.png" % (wdoy)
    plt.savefig(imgname)
    
    
    
    
```
